Creating a basic webscrapper to extract Northern Power grid data. 

```python
import requests
import pandas as pd
import json
url = 'https://power.northernpowergrid.com/Powercut_API/rest/powercuts/getall'
response = requests.get(url)

# if response.status_code == 200:
#     # print response text
#     print(response.text)
```

```python
data = response.json()
print(type(data))
```
