```python
import os
import requests
import time

def get_places(query):
    """
    Retrieves a list of queried points in a given location.

    Args:
        query (str): The location to search for hospitals.

    Returns:
        list: A list of hospital names and addresses.
    """
    url = 'https://maps.googleapis.com/maps/api/place/textsearch/json'
    # key = os.environ.get('GOOGLE_MAPS_API_KEY')
    key = 'AIzaSyAAqqRApAbbg620DzZ1KCk9bOYQtPMf7ss'
    if not key:
        raise ValueError('Google Maps API key not found.')

    params = {
        'query': query,
        'key': key,
    }

    hospitals = []
    while True:
        try:
            res = requests.get(url, params=params)
            res.raise_for_status()
            data = res.json()
            hospitals.extend(data['results'])
            # hospitals.extend(data)
            if 'next_page_token' in data:
                params['pagetoken'] = data['next_page_token']
                time.sleep(2)
            else:
                break
        except:
            "Error"
    return hospitals
```

```python
import requests
import folium
import json

def map_this(query):
    """
    Retrieves a map with all the queried points in a given location.

    Args:
        query (str): The location to search for stored.

    Returns:
        list: A list of hospital names and addresses.
    """
    stored = get_places(query)
    m = folium.Map(location=[54.9783, -1.6178], zoom_start=12) #New Castle
    for i in range(0,len(stored)):
        cc = [stored[i]["geometry"]["location"]["lat"], stored[i]["geometry"]["location"]["lng"]]
        folium.Marker(
                    cc
                    , icon=folium.Icon(icon = 'square', color='red')
                    , tooltip = [stored[i]["name"]] 
                ).add_to(m)
    return m
map_this("All Hospitals in North East England")
```
